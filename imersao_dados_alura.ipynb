## **Desafio Aula 3 | An√°lise salarial de Data Scientists**
An√°lise focada no cargo Data Scientist, calculando o sal√°rio m√©dio anual por pa√≠s e apresentando os resultados em um gr√°fico interativo com Plotly, permitindo uma compara√ß√£o clara entre os pa√≠ses.

```python
# 1. Filtrar os dados para o cargo 'Data Scientist'
df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']

# 2. Agrupar por pa√≠s e calcular a m√©dia salarial
df_salario_por_pais = df_data_scientist.groupby('empresa')['usd'].mean().reset_index()

# 3. Ordenar os resultados para um gr√°fico mais claro (opcional, mas recomendado)
df_salario_por_pais = df_salario_por_pais.sort_values(by='usd', ascending=True)

# 4. Criar o gr√°fico de barras interativo com Plotly
fig = px.bar(
    df_salario_por_pais,
    x='empresa',
    y='usd',
    title='Sal√°rio M√©dio Anual para Data Scientists por Pa√≠s (USD)',
    labels={'empresa': 'Pa√≠s da Empresa', 'usd': 'Sal√°rio M√©dio Anual (USD)'},
    color='empresa', # Colore as barras por pa√≠s
    color_discrete_sequence=px.colors.qualitative.Pastel # Escolhe uma paleta de cores
)

fig.update_layout(xaxis_tickangle=-45) # Inclina os r√≥tulos do eixo X para melhor legibilidade
fig.show()
```

## **Aula 4 | Construindo um dashboard interativo**
Consolida√ß√£o da an√°lise em um dashboard interativo, reunindo gr√°ficos e visualiza√ß√µes para apresenta√ß√£o clara dos dados e apoio √† tomada de decis√£o.

```python
import pycountry
import plotly.express as px

# Fun√ß√£o para converter ISO-2 para ISO-3
def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

# Criar nova coluna com c√≥digo ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcular m√©dia salarial por pa√≠s (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = (
    df_ds
    .groupby('residencia_iso3')['usd']
    .mean()
    .reset_index()
)

# Gerar o mapa
fig = px.choropleth(
    media_ds_pais,
    locations='residencia_iso3',
    color='usd',
    color_continuous_scale='rdylgn',
    title='Sal√°rio m√©dio de Cientista de Dados por pa√≠s',
    labels={
        'usd': 'Sal√°rio m√©dio (USD)',
        'residencia_iso3': 'Pa√≠s'
    }
)

fig.show()
```
